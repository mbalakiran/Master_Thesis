```python
with open('combined_LOD_0_0.csv', 'w', newline='') as f_output:
    csv_output = csv.writer(f_output, delimiter=',',quotechar=' ', quoting=csv.QUOTE_MINIMAL)
    

    for filename in glob.glob('LOD_0_0_*.csv'):
        print (filename)
        
        with open(filename) as file: 
#             fileHandle = open ( 'LOD_0_0_*.csv',"r" )
            lineList = file.readlines()
            file.close()
#             print lineList
#             print "The last line is:"
            a = lineList[len(lineList)-1]
        csv_output.writerow([filename] + [a])
```

```python
df00 = pd.read_csv('combined_LOD_0_0.csv')
df00.to_csv('combined_LOD_0_0.csv',header=["Filename", "Generation", "ID","Score"], index=False)
```

```python
with open('Average_0_0.csv', 'w', newline='') as f_output:
    csv_output = csv.writer(f_output)
    

    for filename in glob.glob('LOD_0_0_*.csv'):
        print (filename)

        with open(filename, newline='') as csvfile:
            averages = []
            reader = csv.DictReader(csvfile)
            values = [ float(row["score"]) for row in reader ]
            avg = sum(values) / len(values)
            averages.append((avg))
            

            for col in zip(*csvfile):
                averages.append(sum(int(x) for x in col) / len(col))

            csv_output.writerow([filename] + averages)
#         csv_output.writerow([filename] + averages)
```
