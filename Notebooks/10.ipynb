```python
def transition_matrix(transitions):
    n = 1+ max(transitions) #number of states

    M = [[0]*n for _ in range(n)]

    for (i,j) in zip(transitions,transitions[1:]):
        M[i][j] += 1

    #now convert to probabilities:
    for row in M:
        s = sum(row)
        if s >= 0:
            row[:] = [f for f in row]
    return M
```

```python
def rm(V):
    pDist = dict()
    inc = 1.0/float(len(V))
    for v in V:
        if v not in pDist:
            pDist[v]=0
        pDist[v]+=inc
    return pDist

rm(rolls)
```
