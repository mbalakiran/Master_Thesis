```python
averages = []
for fname in glob.glob(path):
    with open(filename, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        values = [ float(row["score"]) for row in reader ]
        avg = sum(values) / len(values)
        averages.append((fname, avg))
```

```python
with open('result.csv', 'w', newline='') as f_output:
    csv_output = csv.writer(f_output)
    

    for filename in glob.glob('L*.csv'):
        print (filename)

        with open(filename, newline='') as csvfile:
            averages = []
            reader = csv.DictReader(csvfile)
            values = [ float(row["score"]) for row in reader ]
            avg = sum(values) / len(values)
            averages.append((avg))
            

            for col in zip(*csvfile):
                averages.append(sum(int(x) for x in col) / len(col))

            csv_output.writerow([filename] + averages)
#         csv_output.writerow([filename] + averages)
```
